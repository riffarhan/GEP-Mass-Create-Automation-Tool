```python
# Extract 'Contact' sheet before cleaning
contact_before = input_data.get('Contact', pd.DataFrame())
# Define cleaning functions

def remove_empty_rows(df):
    """
    Remove rows where all elements are NaN or empty.
    """
    cleaned_df = df.dropna(how='all').reset_index(drop=True)
    print(f"Removed empty rows. Remaining rows: {len(cleaned_df)}")
    return cleaned_df

def convert_columns_to_string(df):
    """
    Convert all columns in the DataFrame to strings.
    For numeric columns, ensure integers are represented without decimals.
    """
    for col in df.columns:
        # Convert numeric columns to strings without decimals
        if pd.api.types.is_numeric_dtype(df[col]):
            df[col] = df[col].apply(lambda x: str(int(x)) if pd.notnull(x) and x == x else "")
        else:
            # Convert non-numeric columns directly to strings
            df[col] = df[col].astype(str).replace("nan", "").replace("<NA>", "").fillna("")
    print("Converted all columns to string type.")
    return df


def fill_missing_values(df, fill_value=""):
    """
    Fill missing values in the DataFrame with a default value.
    Replace empty strings as well for consistency.
    """
    df = df.replace({None: fill_value, np.nan: fill_value, "nan": fill_value, "<NA>": fill_value})
    print(f"Filled missing values with: '{fill_value}'")
    return df

def strip_whitespace(df):
    """
    Strip leading and trailing whitespace from string columns.
    """
    str_columns = df.select_dtypes(include=["object", "string"]).columns
    for col in str_columns:
        if df[col].dtype == "object" or pd.api.types.is_string_dtype(df[col]):
            df[col] = df[col].astype(str).str.strip()  # Ensure the column is treated as a string
    print("Stripped whitespace from string columns.")
    return df


def standardize_column_names(df):
    """
    Standardize column names: lowercase, replace spaces with underscores.
    """
    df.columns = df.columns.str.strip()
    df.columns = df.columns.str.replace(" *", "")
    df.columns = df.columns.str.replace("*", "")
    df.columns = df.columns.str.replace("+", "")
    df.columns = df.columns.str.replace(" ", "_")
    df.columns = df.columns.str.replace("(", "")
    df.columns = df.columns.str.replace(")", "")
    df.columns = df.columns.str.replace("|", "")
    df.columns = df.columns.str.replace("nan", "")
    
    print("Standardized column names.")
    return df

def remove_duplicates(df, subset=None):
    """
    Remove duplicate rows based on all columns or specific subset.
    """
    before = len(df)
    deduped_df = df.drop_duplicates(subset=subset)
    after = len(deduped_df)
    print(f"Removed {before - after} duplicate rows.")
    return deduped_df

def validate_data_types(df, expected_types):
    """
    Validate and enforce column data types.
    Parameters:
        - expected_types: dict with column names as keys and expected types as values.
    """
    for column, dtype in expected_types.items():
        if column in df.columns:
            df[column] = df[column].astype(dtype)
    print("Validated and enforced data types.")
    return df

# def filter_invalid_rows(df, column, valid_values):
#     """
#     Remove rows where the specified column has invalid values.
#     """
#     filtered_df = df[df[column].str.endswith(tuple(valid_values), na=False)].reset_index(drop=True)
#     print(f"Filtered invalid rows in column '{column}'. Remaining rows: {len(filtered_df)}")
#     return filtered_df

# Function to clean all sheets in a dictionary
def clean_all_sheets(data_dict, cleaning_functions):
    """
    Apply cleaning functions to all DataFrames in a dictionary.

    Parameters:
    - data_dict: Dictionary of DataFrames to be cleaned.
    - cleaning_functions: List of cleaning functions to apply sequentially.

    Returns:
    - Cleaned dictionary of DataFrames.
    """
    cleaned_data = {}
    for sheet_name, df in tqdm(data_dict.items(), desc="Cleaning Sheets", unit="sheet"):
        print(f"\nCleaning sheet: {sheet_name}")
        for func in cleaning_functions:
            df = func(df)  # Apply each cleaning function
        cleaned_data[sheet_name] = df
    return cleaned_data

# List of cleaning functions to apply
cleaning_functions = [
     convert_columns_to_string,
     remove_empty_rows,
     fill_missing_values,  # Default fill value is ""
     strip_whitespace,
    standardize_column_names
    # lambda df: remove_duplicates(df, subset=None)  # Deduplicate based on all columns
]

# Clean input_data sheets
print("\n=== Cleaning Input Data Sheets ===")
cleaned_input_data = clean_all_sheets(input_data, cleaning_functions)

# Clean master_data sheets
print("\n=== Cleaning Master Data Sheets ===")
cleaned_master_data = clean_all_sheets(master_data, cleaning_functions)

# Extract 'Contact' sheet after cleaning
contact_after = cleaned_input_data.get('Contact', pd.DataFrame())
```

```python
# Helper function: VLOOKUP equivalent
def vlookup(df, lookup_value, lookup_col_index, return_col_index, default=""):
    try:
        match = df.iloc[:, lookup_col_index].eq(lookup_value)
        if match.any():
            return df.iloc[match.idxmax(), return_col_index]
        else:
            return default
    except Exception as e:
        return default

# Helper function: Split text by lengths
def split_text(text, lengths):
    if not isinstance(text, str):
        text = str(text)
    text = text.strip()
    parts = []
    start = 0
    for length in lengths:
        part = text[start:start + length].strip()
        parts.append(part)
        start += length
    while len(parts) < len(lengths):
        parts.append('')
    return parts

# Transform function
def transform_user_basic(user_basic, master_data):
    # Create a new instance of the DataFrame
    user_basicdetails = user_basic.copy()

    # Map EngagementModelId-LOB
    user_basicdetails['EngagementModelId-LOB'] = user_basicdetails['LOB'].map(
        lambda x: vlookup(master_data['master_org_entity'], x.strip() if x else '', 3, 6)
    )

    # Map AccountGroupDesc
    user_basicdetails['AccountGroupDesc'] = user_basicdetails['AccountGroup'].map(
        lambda x: vlookup(master_data['master_onboarding_type'], x.strip() if x else '', 4, 5)
    )

    # Generate Full Name
    user_basicdetails['Full Name'] = user_basicdetails.apply(
        lambda row: f"{row['Name']}, {row['Prefix']}".strip(', '), axis=1
    )

    # Count Full Name length
    user_basicdetails['Count Full Name'] = user_basicdetails['Full Name'].apply(len)

    # Split Full Name into parts
    user_basicdetails[['Name1', 'Name2', 'Name3', 'Name4']] = pd.DataFrame(
        user_basicdetails['Full Name'].apply(lambda x: split_text(x, [35, 35, 35, 35])).tolist(),
        index=user_basicdetails.index
    )

    # Map Check Prefix
    user_basicdetails['Master Check Prefix'] = user_basicdetails['Prefix'].map(
        lambda x: vlookup(master_data['master_supplier_name_prefix'], x.strip() if x else '', 2, 2)
    )
    # Step 1: Filter master_category to ensure unique mapping based on the lowest ClientCategoryCode
    # Sort by CategoryName and ClientCategoryCode (ascending order) and drop duplicates
    master_category_unique = (
        master_category.sort_values(by=['CategoryName', 'ClientCategoryCode'], ascending=True)
        .drop_duplicates(subset='CategoryName', keep='first')
    )
    
    # Create a mapping from CategoryName to CategoryCode, now considering only the lowest ClientCategoryCode
    category_mapping = master_category_unique.set_index('CategoryName')['CategoryCode'].to_dict()
    
    # Normalize category_mapping keys to lowercase for case-insensitive matching
    category_mapping_normalized = {key.lower(): value for key, value in category_mapping.items()}
    
    # Step 2: Map SubCategoryId1
    user_basicdetails['SubCategoryId1'] = user_basicdetails['Subcategory'].apply(
        lambda x: category_mapping_normalized.get(x.split(',')[0].strip().lower(), '') 
        if isinstance(x, str) and ',' in x 
        else category_mapping_normalized.get(x.strip().lower(), '') 
        if isinstance(x, str) 
        else ''
    )
    
    # Step 3: Map SubCategoryId2
    user_basicdetails['SubCategoryId2'] = user_basicdetails['Subcategory'].apply(
        lambda x: category_mapping_normalized.get(x.split(',')[1].strip().lower(), '') 
        if isinstance(x, str) and ',' in x and len(x.split(',')) > 1 
        else ''
    )
    # Concatenate SubCategoryId1 and SubCategoryId2 into SubCategoryIdResult
    user_basicdetails['SubCategoryIdResult'] = user_basicdetails.apply(
        lambda row: ','.join(filter(None, [row['SubCategoryId1'].strip(), row['SubCategoryId2'].strip()])), 
        axis=1
    )

    
    # Map ServiceLocationId
    user_basicdetails['Master ServiceLocationId'] = user_basicdetails['ServiceLocation'].map(
        lambda x: vlookup(master_data['master_region'], x.strip() if x else '', 2, 3)
    )

    # Map Check Classification
    user_basicdetails['Master Classification'] = user_basicdetails['Classification'].map(
        lambda x: vlookup(master_data['master_classification'], x.strip() if x else '', 4, 4)
    )

    # Map Check Ctrl.Department
    user_basicdetails['Master Ctrl.Department'] = user_basicdetails['ControlDepartment'].map(
        lambda x: vlookup(master_data['master_control_department'], x.strip() if x else '', 4, 4)
    )

    # Map Check Currency
    user_basicdetails['Master Currency'] = user_basicdetails['Currency'].map(
        lambda x: vlookup(master_data['master_currency'], x.strip() if x else '', 2, 2)
    )

    # Generate Full Address
    user_basicdetails['AddressFull'] = user_basicdetails[['Address1', 'Address2']].fillna('').apply(
        lambda x: ' '.join(x).strip(), axis=1
    )

    # Count Full Address length
    user_basicdetails['Count Full Address'] = user_basicdetails['AddressFull'].apply(len)

# Split AddressFull into parts with temporary unique column names
    address_split = pd.DataFrame(
    user_basicdetails['AddressFull'].apply(lambda x: split_text(x, [35, 35, 35, 35, 35])).tolist(),
    index=user_basicdetails.index,
    columns=['Master Address1', 'Master Address2', 'Master Address3', 'Master Address4', 'TempAddress5']
)

    # Add the split address columns to the DataFrame
    user_basicdetails = pd.concat([user_basicdetails, address_split], axis=1)


    # Drop the original Address1 and Address2 columns if they are no longer needed
    # user_basicdetails.drop(columns=['TempAddress1', 'TempAddress2'], inplace=True, errors='ignore')


    # Map StateId
    user_basicdetails['Master StateId'] = user_basicdetails['State'].map(
        lambda x: vlookup(master_data['master_state'], x.strip() if x else '', 3, 3)
    )

    # Map TradingPartnerGEP
    user_basicdetails['Master TradingPartnerGEP'] = user_basicdetails['TradingPartner'].map(
        lambda x: vlookup(master_data['master_trading_partner'], x.strip() if x else '', 2, 3)
    )

    # Map InstructionKeyGEP
    user_basicdetails['Master InstructionKeyGEP'] = user_basicdetails['InstructionKey'].map(
        lambda x: vlookup(master_data['master_instruction_key'], x.strip() if x else '', 2, 5)
    )

    # Copy LocationCode and LocationName from VendorCode
    user_basicdetails['LocationCode'] = user_basicdetails['VendorCode']
    user_basicdetails['LocationName'] = user_basicdetails['VendorCode']
    
    # Assuming column indices: U (20th), V (21st), W (22nd), X (23rd)
    # Adjust indices based on zero-based indexing for pandas DataFrame
    column_indices = {
        'NewVatReg': 20,  # U
        'NewNIK': 21,     # V
        'NewNITKU': 22,   # W
        'NewVatLama': 23  # X
    }
    
    # Reorder columns to move the specified columns to the end
    # user_basicdetails = user_basicdetails[[col for col in user_basicdetails.columns if col not in columns_to_move] + columns_to_move]
    user_basicdetails.rename(
        columns={
            'NewVatReg': 'VatReg',
            'NewNIK': 'NIK',
            'NewNITKU': 'NITKU',
            'NewVatLama': 'VatLama'
        },
        inplace=True
    )
        # Validation logic
    columns_to_compare = ['Prefix', 'Classification', 'ControlDepartment', 'Currency']
    master_columns = ['Master Check Prefix',  'Master Classification',
                      'Master Ctrl.Department', 'Master Currency']


    
    # Map StateId
    user_basicdetails['Master StateId'] = user_basicdetails['State'].map(
        lambda x: vlookup(master_data['master_state'], x.strip() if x else '', 4, 3)
    )

    # Map TradingPartnerGEP
    user_basicdetails['Master TradingPartnerGEP'] = user_basicdetails['TradingPartner'].map(
        lambda x: vlookup(master_data['master_trading_partner'], x.strip() if x else '', 2, 3)
    )

    # Map InstructionKeyGEP
    user_basicdetails['Master InstructionKeyGEP'] = user_basicdetails['InstructionKey'].map(
        lambda x: vlookup(master_data['master_instruction_key'], x.strip() if x else '', 2, 5)
    )
    # Convert columns to string for comparison
    for col in columns_to_compare + master_columns:
        user_basicdetails[col] = user_basicdetails[col].fillna('').astype(str).str.strip()

    # Find mismatches
    mismatches = []
    for col, master_col in zip(columns_to_compare, master_columns):
        mismatch_rows = user_basicdetails[
            (user_basicdetails['VendorCode'] != '') &  # Skip rows with blank VendorCode
            (user_basicdetails[col].str.casefold() != user_basicdetails[master_col].str.casefold())
        ]
        for _, row in mismatch_rows.iterrows():
            mismatches.append({
                'VendorCode': row['VendorCode'],
                'Column': col,
                'OriginalValue': row[col],
                'MasterValue': row[master_col]
            })

    mismatch_df = pd.DataFrame(mismatches)

    return user_basicdetails, mismatch_df

# Apply the function
transformed_basic, mismatches = transform_user_basic(user_basic, {
    'master_org_entity': master_org_entity,
    'master_onboarding_type': master_onboarding_type,
    'master_supplier_name_prefix': master_supplier_name_prefix,
    'master_category': master_category,
    'master_region': master_region,
    'master_classification': master_classification,
    'master_control_department': master_control_department,
    'master_currency': master_currency,
    'master_state': master_state,
    'master_trading_partner': master_trading_partner,
    'master_instruction_key': master_instruction_key,
}) 
# Output results
if not mismatches.empty:
    print("\n=== Mismatches Found ===")
    print(mismatches)
else:
    print("\nNo mismatches found.")
transformed_basic['VendorCode'] = transformed_basic['VendorCode'].str.strip()
transformed_basic.replace("", pd.NA, inplace=True)
transformed_basic.dropna(subset=["VendorCode"], how='all', inplace=True)
```

```python
# Helper Function: VLOOKUP Equivalent
def vlookup(df, lookup_value, lookup_col_index, return_col_index, default=""):
    try:
        match = df.iloc[:, lookup_col_index].eq(lookup_value)
        if match.any():
            return df.iloc[match.idxmax(), return_col_index]
        else:
            return default
    except Exception as e:
        return default

def vlookup_contains(df, lookup_value, lookup_col_index, return_col_index, default=""):
    try:
        # Perform a substring match instead of exact match
        matches = df.iloc[:, lookup_col_index].apply(lambda x: lookup_value in str(x))
        if matches.any():
            return df.iloc[matches.idxmax(), return_col_index]  # Return the first match
        else:
            return default
    except Exception as e:
        return default


# Transform Function
def transform_user_company(user_company, user_basic, master_data):
    """
    Transforms the `user_company` DataFrame based on Excel logic.
    
    Parameters:
        user_company (pd.DataFrame): Original company data.
        user_basic (pd.DataFrame): Basic user data for lookup.
        master_data (dict): Dictionary of all master data tables.
    
    Returns:
        pd.DataFrame: Transformed data.
        pd.DataFrame: Mismatch details.
    """
    # Create a copy of the input DataFrame
    transformed = user_company.copy()

    # Apply transformations and add new columns
    transformed['Master LocationCode'] = transformed['VendorCode'].map(
        lambda x: vlookup(user_basic, x.strip() if pd.notnull(x) else '', 1, 1)
    )

    transformed['Master Engagement ModelId-Cocode'] = transformed['CompanyCode'].map(
        lambda x: vlookup(master_data['master_org_entity'], x.strip() if pd.notnull(x) else '', 3, 6)
    )

    transformed['Master Entity Type'] = transformed['CompanyCode'].map(
        lambda x: vlookup(master_data['master_org_entity'], x.strip() if pd.notnull(x) else '', 3, 1)
    )

    transformed['Master LOB Code'] = transformed['CompanyCode'].map(
        lambda x: vlookup(master_data['master_org_entity'], x.strip() if pd.notnull(x) else '', 3, 0)
    )
    
    transformed['Master AccountGroup'] = transformed['AccountGroup'].map(
        lambda x: vlookup(master_data['master_onboarding_type'], x.strip() if pd.notnull(x) else '', 4, 5)
    )

    transformed['Master ReconAccount'] = transformed['ReconAccount'].map(
        lambda x: vlookup(master_data['master_recon_account'], x.strip() if pd.notnull(x) else '', 2, 4)
    )

    transformed['Master SortKey'] = transformed['SortKey'].map(
        lambda x: vlookup(master_data['master_sort_key'], x.strip() if pd.notnull(x) else '', 2, 2)
    )

    transformed['Master CashMngmtGroup'] = transformed['CashMngmtGroup'].map(
        lambda x: vlookup(master_data['master_planning_cash_management'], x.strip() if pd.notnull(x) else '', 2, 4)
    )

    transformed['Master PaymentMethod'] = transformed['PaymentMethod'].map(
        lambda x: vlookup(master_data['master_payment_method'], x.strip() if pd.notnull(x) else '', 2, 5)
    )

    # Concatenate WHTType + WHTCode
    transformed['Master WHTType+Code'] = transformed['WHTType'].fillna('') + transformed['WHTCode'].fillna('')


    master_wht['SubTax'] = master_wht.apply(
        lambda row: f"{row['Sub_Tax_Type']}{row['Tax_Code']}".strip(), axis=1
    )
    sub_tax_index = master_wht.columns.get_loc('SubTax')

    # Tax Details from `master_wht`
    for col, master_col in [
        ('Master taxTypeCode', 7),
        ('Master taxTypeName', 7),
        ('Master taxCodeName', 2),
        ('Master taxCodeCode', 2),
        ('Master subTypeName', 8),
        ('Master subTypeCode', 8),
        ('Master taxDescription', 14),
        ('Master taxPercentage', 9),
    ]:
        transformed[col] = transformed['Master WHTType+Code'].map(
            lambda x: vlookup(master_data['master_wht'], x.strip() if pd.notnull(x) else '', sub_tax_index, master_col)
        )
    transformed['Master PaymentTerms'] = transformed['PaymentTerms'].map(
        lambda x: vlookup_contains(master_data['master_payment_terms'], x.strip() if pd.notnull(x) else '', 3, 3)
    )

    columns_to_add = [
        'paymentTermRowNumber', 'withholdingTaxId', 'incoTermId',
        'lobDetailsRowNumber', 'companyDataExtensionRowNumber',
        'purchasingDataExtensionRowNumber', 'reconciliationAccountRowNumber'
    ]
    for column in columns_to_add:
        transformed[f"Added{column}"] = ""

    # Define a mapping for the columns to populate based on the 'Entity Type'
    column_mapping = {
        'Company Code': ['withholdingTaxId', 'companyDataExtensionRowNumber', 'reconciliationAccountRowNumber'],
        'Purchasing Org': ['paymentTermRowNumber', 'incoTermId', 'purchasingDataExtensionRowNumber']
    }

    # Populate the Added columns based on Entity Type
    for entity_type, columns in column_mapping.items():
        entity_filter = transformed['Master Entity Type'] == entity_type
        for column in columns:
            transformed.loc[entity_filter, f"Added{column}"] = transformed.loc[entity_filter, 'Master Engagement ModelId-Cocode']

    # Validation Logic
    columns_to_compare = [
        'AccountGroup', 'ReconAccount',
        'SortKey', 'CashMngmtGroup', 'PaymentMethod', 'PaymentTerms'
    ]
    master_columns = [
        'Master AccountGroup', 'Master ReconAccount',
        'Master SortKey', 'Master CashMngmtGroup', 'Master PaymentMethod', 'Master PaymentTerms'
    ]

    for col in columns_to_compare + master_columns:
        transformed[col] = transformed[col].fillna('').astype(str).str.strip()

    # Find mismatches
    mismatch_records = []
    for col, master_col in zip(columns_to_compare, master_columns):
        if master_col in transformed.columns:
            mismatches = transformed[
                (transformed['VendorCode'] != '') &  # Skip rows with blank VendorCode
                (~transformed.apply(
                    lambda row: row[col].casefold() in row[master_col].casefold(), axis=1  # Substring match
                ))
            ]
            for _, row in mismatches.iterrows():
                mismatch_records.append({
                    'VendorCode': row['VendorCode'],
                    'Column': col,
                    'OriginalValue': row[col],
                    'MasterValue': row[master_col]
                })
    # Create a DataFrame for mismatches
    mismatch_df = pd.DataFrame(mismatch_records)

    return transformed, mismatch_df

# Apply the transformation
transformed_user_company, mismatches = transform_user_company(
    user_company=user_company,
    user_basic=user_basic,
    master_data={
        'master_org_entity': master_org_entity,
        'master_onboarding_type': master_onboarding_type,
        'master_recon_account': master_recon_account,
        'master_sort_key': master_sort_key,
        'master_planning_cash_management': master_planning_cash_management,
        'master_payment_method': master_payment_method,
        'master_wht': master_wht,
        'master_payment_terms': master_payment_terms,
    }
)

# Output mismatches for validation
if not mismatches.empty:
    print("\n=== Mismatches Found ===")
    print(mismatches)
else:
    print("\nNo mismatches found.")


transformed_user_company['VendorCode'] = transformed_user_company['VendorCode'].str.strip()
transformed_user_company.replace("", pd.NA, inplace=True)
transformed_user_company.dropna(subset=["VendorCode"], how='all', inplace=True)
```

```python
# Define helper functions
def vlookup(df, lookup_value, lookup_col_index, return_col_index, default=""):
    try:
        match = df.iloc[:, lookup_col_index].eq(lookup_value)
        if match.any():
            return df.iloc[match.idxmax(), return_col_index]
        else:
            return default
    except Exception as e:
        return default

def mod_row(row_index, mod_base=9):
    return (row_index - 2) % mod_base + 1

# Process user_contact
def transform_user_contact(user_contact, user_basic):
    user_contact_transformed = user_contact.copy()

    # Vendor Code (VLOOKUP from user_basic sheet)
    user_contact_transformed['Vendor Code'] = user_contact_transformed['VendorCode'].map(
        lambda x: vlookup(user_basic, x.strip() if pd.notnull(x) else '', 1, 1)
    )

    # First Name
    user_contact_transformed['First Name'] = user_contact_transformed['FirstName'].apply(
        lambda x: "-" if pd.isna(x) or not str(x).strip() else str(x).strip().upper()
    )

    user_contact_transformed['Last Name'] = user_contact_transformed['LastName'].apply(
        lambda x: "-" if pd.isna(x) or x is pd.NA or not str(x).strip() else str(x).strip().upper()
    )


    # EmailAddress
    user_contact_transformed['EmailAddress'] = user_contact_transformed['emailAddress'].apply(
        lambda x: str(x).strip().upper() if pd.notnull(x) else ""
    )

    # Add RowUnique column with numbers cycling from 1 to 9
    user_contact_transformed['RowUnique'] = user_contact_transformed.index.map(lambda x: (x % 9) + 1)

    # ClientContactCode
    user_contact_transformed['ClientContactCode'] = user_contact_transformed.apply(
        lambda row: f"{row['VendorCode']}{row['RowUnique']}", axis=1
    )

    return user_contact_transformed

# Transform the Contact data
transformed_contact = transform_user_contact(user_contact, user_basic)

transformed_contact['VendorCode'] = transformed_contact['VendorCode'].str.strip()
transformed_contact.replace("", pd.NA, inplace=True)
transformed_contact.dropna(subset=["VendorCode"], how='all', inplace=True)

# Display the transformed data
transformed_contact
```

## Batch

```python
# Define the new folder name
new_folder_name = "Batch_Output"

# Combine the paths to create the full path for the new folder
output_folder_batch = os.path.join(output_folder, new_folder_name)
```

```python
# Ensure the new folder exists
if not os.path.exists(output_folder_batch):
    os.makedirs(output_folder_batch)
    print(f"New folder created at: {output_folder_batch}")
else:
    print(f"Folder already exists at: {output_folder_batch}")

# Define batch sizes: first 1, then 10, 50, 100; default for subsequent batches is 350
batch_sizes = [1, 10, 50, 100]
default_batch_size = 350

# Ensure "Partners" exists in the dataframes dictionary
if 'Partners' in dataframes:
    partners_df = dataframes['Partners']

    # Ensure "State Code" is treated as a string
    if 'State Code' in partners_df.columns:
        partners_df['State Code'] = partners_df['State Code'].astype(str)

    start_index = 0
    batch_index = 1  # Batch numbering

    while start_index < len(partners_df):
        # Determine current batch size based on our sequence or use default afterward
        if batch_index <= len(batch_sizes):
            current_batch_size = batch_sizes[batch_index - 1]
        else:
            current_batch_size = default_batch_size

        # Define the batch file name
        batch_file_name = os.path.join(
            output_folder_batch,
            f"Mass Create Data Batch {batch_index}.xlsx"
        )

        # Get the chunk of rows
        end_index = min(start_index + current_batch_size, len(partners_df))
        partners_chunk = partners_df.iloc[start_index:end_index]

        # Write the batch to an Excel file
        with pd.ExcelWriter(batch_file_name, engine='xlsxwriter') as batch_writer:
            # Save the chunk of "Partners" data
            partners_chunk.to_excel(batch_writer, sheet_name='Partners', index=False)

            # Iterate through other dataframes for mapping
            for sheet_name, df in dataframes.items():
                if sheet_name != 'Partners':
                    # Filter the DataFrame based on the unique values from "Partners"
                    filtered_df = df[df.iloc[:, 0].isin(partners_chunk.iloc[:, 0].unique())]
                    # Write the filtered DataFrame to the batch Excel file if it contains data
                    if not filtered_df.empty:
                        filtered_df.to_excel(batch_writer, sheet_name=sheet_name, index=False)

        print(f"Batch {batch_index} saved successfully with {len(partners_chunk)} rows to: {batch_file_name}")

        # Update the start index and increment batch index
        start_index = end_index
        batch_index += 1

else:
    print("Error: 'Partners' sheet not found in the data.")
```
